```python
#start - start value
#gradient_func - the gradient_function
#learning_rate
#max_iterations - a limit on how many jumps we want
#tolerance - the value to stop the algo at(how close to minima)

def gradient_descent(start, gradient_func, learning_rate, max_iterations, tolerance = 0.01):
    steps = np.array([start])
    
    x = start
    
    for _ in range(max_iterations):
        diff = learning_rate * gradient_func(x)
        
        if np.abs(diff) < tolerance:
            break
        x = x - diff
        steps = np.append(steps, x)
    return steps, x
```

```python
#example 1
#f(x) = x^2+4x+1

def ex_1(x):
    return x**2+4*x+1
```

```python
def g_ex_1(x):
    return 2*x+4
```

```python
history, result = gradient_descent(9, g_ex_1, 0.1, 100)
```
